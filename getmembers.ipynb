```python
import csv, requests, xml.etree.ElementTree as ET
import xmltodict, pandas as pd, numpy as np
```

```python
def loadMembers():
    url = "https://www.ourcommons.ca/Members/en/search/XML"
    resp = requests.get(url)
    with open('parliamentMembers.xml', 'wb') as f:
        f.write(resp.content)
```

```python
loadMembers()
```

```python
with open('parliamentMembers.xml', encoding='utf-8') as fd:
    doc = xmltodict.parse(fd.read())

members_list = doc['ArrayOfMemberOfParliament']['MemberOfParliament']

df = pd.DataFrame(members_list)
df
```

```python
df = df.fillna(value=np.nan)
df
```
