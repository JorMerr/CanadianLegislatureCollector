```python
import csv, requests, xml.etree.ElementTree as ET
```

```python
def loadMembers():
    url = "https://www.ourcommons.ca/Members/en/search/XML"
    resp = requests.get(url)
    with open('parliamentMembers.xml', 'wb') as f:
        f.write(resp.content)
```

```python
tree = ET.parse('parliamentMembers.xml')
root = tree.getroot()

for member in root.iter('PersonOfficialFirstName'):
    print(member.attrib)
```

```python
def parseXML(xmlfile):
    tree = ET.parse(xmlfile)
    root = tree.getroot()
    members = []

    for item in root.findall('MemberOfParliament'):
        membersInfo = {}
        for child in item:
            if child.tag == 'PersonOfficialFirstName':
                members['MemberFirstName'] = child.attrib['PersonOfficialFirstName']


        members.append(membersInfo)

    return members
```
